```python
- 1.2 Ôn bài cũ — 3 câu hỏi nhanh

• Q1: Sự khác biệt giữa loc[] và iloc[]?
  - loc[] (Label-based): Định vị dữ liệu dựa vào tên nhãn (label) của dòng hoặc cột.
    - Ví dụ: df.loc[0, 'Name'] -> Tìm dòng có nhãn là 0 và cột có tên là 'Name'.
    - Đặc điểm cắt lát (Slicing): Lấy bao gồm cả phần tử cuối (Ví dụ: 1:3 sẽ lấy cả dòng 1, 2, 3).
  - iloc[] (Integer position-based): Định vị dữ liệu dựa vào chỉ số vị trí nguyên (index) tương tự như mảng trong Python.
    - Ví dụ: df.iloc[0, 1] -> Tìm dòng đầu tiên (vị trí 0) và cột thứ hai (vị trí 1).
    - Đặc điểm cắt lát (Slicing): Không bao gồm phần tử cuối (Ví dụ: 1:3 chỉ lấy dòng 1 và 2).

• Q2: Khi nào dùng fillna(median) thay vì fillna(mean)?
  - Dùng fillna(mean) (Giá trị trung bình): Khi dữ liệu phân phối chuẩn, đối xứng và không có giá trị ngoại lai (outliers).
  - Dùng fillna(median) (Giá trị trung vị): Khi dữ liệu bị lệch (skewed) hoặc có chứa các giá trị ngoại lai quá lớn hoặc quá nhỏ.
    - Lý do: Giá trị trung bình (mean) rất dễ bị kéo lệch bởi các số liệu cực đoan, trong khi trung vị (median) giữ được xu hướng trung tâm chính xác hơn của dữ liệu thực tế.



[Image of skewed distribution mean median mode]


• Q3: drop_duplicates() mặc định giữ bản sao nào?
  - Mặc định, hàm này có tham số keep='first'.
  - Ý nghĩa: Nó sẽ giữ lại bản ghi đầu tiên xuất hiện trong dữ liệu và xóa bỏ tất cả các bản sao trùng lặp ở phía sau.

---

- Điểm trọng tâm cần nắm

• Boolean indexing: & | ~ — không dùng and, or, not
  - Khi lọc dữ liệu có điều kiện trong Pandas, bắt buộc phải dùng các toán tử bitwise (theo từng phần tử):
    - & thay cho and (Phép VÀ)
    - | thay cho or (Phép HOẶC)
    - ~ thay cho not (Phép PHỦ ĐỊNH)
  - Lý do: Các từ khóa and, or, not của Python chỉ kiểm tra tính đúng sai của toàn bộ một đối tượng (bảng dữ liệu). Trong khi đó, Pandas cần so sánh và lọc từng dòng một (element-wise).
  - Mẹo: Luôn nhớ bọc các điều kiện nhỏ trong dấu ngoặc đơn ().
    - Ví dụ đúng: df[(df['Tuổi'] > 20) & (df['Điểm'] > 5)]
```
